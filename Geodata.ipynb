```python
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
import pandas as pd
import numpy as np

# generate the clusters of datas
data = pd.read_csv("Greece_EQ_Euro_wDS_AIR_CT19.csv")
quake_map = folium.Map(location=[data['Latitude'].mean(), 
                                data['Longitude'].mean()], 
                                zoom_start=4)
quake_map.add_child(FastMarkerCluster(data[['Latitude', 'Longitude']].values.tolist()))
quake_ma
```

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon


info = pd.read_csv('Cubeinfo.csv')
geodata = gpd.GeoDataFrame()
geodata['cube_ID']= info['cube_ID']
geodata['geometry'] = None
geodata['num_events']= info['num_events']
geodata['loss_sum']= info['loss_sum']

x=19
r=1
i=0
while x<31:
    y=34
    while y<43:
        geodata.loc[i,'geometry']  = Polygon([(x,y),(x,y+r),(x+r,y+r),(x+r,y)])
        y+=r
        i+=1
    x+=r
print(i)
geodata = geodata.to_json()

m = folium.Map(location=[38, 25], zoom_start=5)

folium.GeoJson(data=geodata).add_to(m)
m
```

```python
m = folium.Map(location=[36, 30], zoom_start=5)

choropleth = folium.Choropleth(
    geo_data=geodata,
    name="choropleth",
    data=info,
    columns=["cube_ID", "num_events"],
    key_on='feature.properties.cube_ID',
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Cell events",
).add_to(m)

choropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(['num_events'], labels=False))
m
```

```python
m = folium.Map(location=[38, 25], zoom_start=5)

choropleth = folium.Choropleth(
    geo_data=geodata,
    name="choropleth",
    data=info,
    columns=["cube_ID", "loss_sum"],
    key_on='feature.properties.cube_ID',
    fill_color="YlGn",
    fill_opacity=0.5,
    line_opacity=0.2,
    legend_name="Cell Risk",
).add_to(m)

choropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(['loss_sum'], labels=False))
m
```
